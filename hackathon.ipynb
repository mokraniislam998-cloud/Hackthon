```python
import csv
import time
import requests
import pandas as pd
from pathlib import Path

# =========================
# CONFIG
# =========================
API_KEY = "jklk"

# Endpoint PRIM
STOP_MONITORING_URL = "https://prim.iledefrance-mobilites.fr/marketplace/stop-monitoring"

# Jeu de données open data qui liste le périmètre temps réel disponible
DATASET_ID = "perimetre-des-donnees-temps-reel-disponibles-sur-la-plateforme-dechanges-stif"
PERIMETER_CSV_URL = (
    f"https://data.iledefrance.fr/api/explore/v2.1/catalog/datasets/"
    f"{DATASET_ID}/exports/csv"
)

# Respect du quota : adapte ces valeurs à ton compte
MAX_CALLS_PER_RUN = 500          # nombre max d'appels pour ce lancement
SECONDS_BETWEEN_CALLS = 0.30     # ~3,3 req/s
TIMEOUT = 20

OUT_DIR = Path("idfm_exports")
OUT_DIR.mkdir(exist_ok=True)

PERIMETER_FILE = OUT_DIR / "perimetre_temps_reel.csv"
OUTPUT_FILE = OUT_DIR / "stop_monitoring_passages.csv"
ERROR_FILE = OUT_DIR / "stop_monitoring_errors.csv"


# =========================
# OUTILS
# =========================
def download_perimeter_csv():
    print("Téléchargement du périmètre temps réel...")
    r = requests.get(PERIMETER_CSV_URL, timeout=TIMEOUT)
    r.raise_for_status()
    PERIMETER_FILE.write_bytes(r.content)
    print(f"Fichier périmètre sauvegardé : {PERIMETER_FILE}")


def find_column(df, candidates):
    """
    Cherche une colonne par nom probable.
    """
    normalized = {c.lower().strip(): c for c in df.columns}
    for cand in candidates:
        key = cand.lower().strip()
        if key in normalized:
            return normalized[key]

    # recherche plus souple
    for c in df.columns:
        lc = c.lower().strip()
        for cand in candidates:
            if cand.lower() in lc:
                return c
    return None


def load_monitoring_refs():
    """
    Charge les MonitoringRef depuis le CSV du périmètre.
    Le schéma exact peut varier, donc on cherche plusieurs noms possibles.
    """
    df = pd.read_csv(PERIMETER_FILE, sep=None, engine="python")

    # Colonnes probables selon les variantes du dataset
    stop_point_col = find_column(df, [
        "monitoringref",
        "monitoring_ref",
        "stop_point_id",
        "id_ref_zde",
        "zdeid_ref",
        "zdeid",
        "stop_id",
        "id_arret",
    ])

    stop_area_col = find_column(df, [
        "stop_area_id",
        "id_ref_zda",
        "zdaid_ref",
        "zdaid",
        "zone_id",
    ])

    line_col = find_column(df, [
        "lineref",
        "line_ref",
        "codifligne",
        "line_id",
        "id_ligne",
    ])

    if not stop_point_col and not stop_area_col:
        raise ValueError(
            f"Impossible de trouver une colonne d'identifiant d'arrêt. "
            f"Colonnes présentes: {list(df.columns)}"
        )

    refs = []

    for _, row in df.iterrows():
        monitoring_ref = None
        line_ref = None

        if stop_point_col and pd.notna(row.get(stop_point_col)):
            value = str(row[stop_point_col]).strip()
            if value:
                # si la valeur n'est pas déjà au format STIF
                monitoring_ref = value if value.startswith("STIF:") else f"STIF:StopPoint:Q:{value}:"

        if not monitoring_ref and stop_area_col and pd.notna(row.get(stop_area_col)):
            value = str(row[stop_area_col]).strip()
            if value:
                monitoring_ref = value if value.startswith("STIF:") else f"STIF:StopArea:SP:{value}:"

        if line_col and pd.notna(row.get(line_col)):
            lv = str(row[line_col]).strip()
            if lv:
                line_ref = lv if lv.startswith("STIF:") else f"STIF:Line::{lv}:"

        if monitoring_ref:
            refs.append((monitoring_ref, line_ref))

    # dédoublonnage
    refs = list(dict.fromkeys(refs))
    print(f"{len(refs)} couples (MonitoringRef, LineRef) chargés")
    return refs


def dig(obj, *path, default=None):
    cur = obj
    for p in path:
        if isinstance(cur, dict):
            cur = cur.get(p)
        else:
            return default
        if cur is None:
            return default
    return cur


def extract_monitored_stop_visits(payload):
    """
    Essaie d'extraire les visites depuis une réponse SIRI Lite JSON.
    """
    # chemin le plus courant
    visits = dig(payload, "Siri", "ServiceDelivery", "StopMonitoringDelivery")
    if isinstance(visits, list) and visits:
        visits = visits[0].get("MonitoredStopVisit", [])
        if isinstance(visits, list):
            return visits

    # autres cas possibles
    if isinstance(payload, dict):
        if "MonitoredStopVisit" in payload and isinstance(payload["MonitoredStopVisit"], list):
            return payload["MonitoredStopVisit"]

    return []


def flatten_visit(visit, requested_monitoring_ref, requested_line_ref):
    """
    Aplatit une entrée MonitoredStopVisit vers une ligne CSV.
    """
    mvj = visit.get("MonitoredVehicleJourney", {}) if isinstance(visit, dict) else {}

    monitored_call = mvj.get("MonitoredCall", {}) if isinstance(mvj, dict) else {}
    monitored_stop = mvj.get("MonitoredStopVisit", {}) if isinstance(mvj, dict) else {}

    return {
        "requested_monitoring_ref": requested_monitoring_ref,
        "requested_line_ref": requested_line_ref or "",
        "response_timestamp": visit.get("RecordedAtTime", ""),
        "line_ref": mvj.get("LineRef", ""),
        "direction_name": mvj.get("DirectionName", ""),
        "published_line_name": mvj.get("PublishedLineName", ""),
        "operator_ref": mvj.get("OperatorRef", ""),
        "destination_name": mvj.get("DestinationName", ""),
        "monitoring_ref": monitored_call.get("StopPointRef", "") or requested_monitoring_ref,
        "stop_name": monitored_call.get("StopPointName", ""),
        "visit_number": visit.get("ItemIdentifier", ""),
        "vehicle_ref": mvj.get("VehicleRef", ""),
        "dated_vehicle_journey_ref": mvj.get("DatedVehicleJourneyRef", ""),
        "expected_arrival_time": monitored_call.get("ExpectedArrivalTime", ""),
        "expected_departure_time": monitored_call.get("ExpectedDepartureTime", ""),
        "aimed_arrival_time": monitored_call.get("AimedArrivalTime", ""),
        "aimed_departure_time": monitored_call.get("AimedDepartureTime", ""),
        "arrival_status": monitored_call.get("ArrivalStatus", ""),
        "departure_status": monitored_call.get("DepartureStatus", ""),
    }


def call_stop_monitoring(monitoring_ref, line_ref=None):
    headers = {"apikey": API_KEY}
    params = {"MonitoringRef": monitoring_ref}
    if line_ref:
        params["LineRef"] = line_ref

    r = requests.get(
        STOP_MONITORING_URL,
        headers=headers,
        params=params,
        timeout=TIMEOUT,
    )
    return r


# =========================
# MAIN
# =========================
def main():
    if not PERIMETER_FILE.exists():
        download_perimeter_csv()

    refs = load_monitoring_refs()

    rows = []
    errors = []
    calls_done = 0

    for monitoring_ref, line_ref in refs:
        if calls_done >= MAX_CALLS_PER_RUN:
            print("Limite d'appels atteinte pour ce run.")
            break

        try:
            resp = call_stop_monitoring(monitoring_ref, line_ref)
            calls_done += 1

            if resp.status_code != 200:
                errors.append({
                    "monitoring_ref": monitoring_ref,
                    "line_ref": line_ref or "",
                    "status_code": resp.status_code,
                    "body": resp.text[:1000],
                })
            else:
                payload = resp.json()
                visits = extract_monitored_stop_visits(payload)

                if not visits:
                    errors.append({
                        "monitoring_ref": monitoring_ref,
                        "line_ref": line_ref or "",
                        "status_code": 200,
                        "body": "Aucun passage trouvé ou structure JSON inattendue",
                    })
                else:
                    for visit in visits:
                        rows.append(flatten_visit(visit, monitoring_ref, line_ref))

        except Exception as e:
            errors.append({
                "monitoring_ref": monitoring_ref,
                "line_ref": line_ref or "",
                "status_code": "EXCEPTION",
                "body": str(e),
            })

        time.sleep(SECONDS_BETWEEN_CALLS)

    # Sauvegarde CSV résultats
    if rows:
        df_rows = pd.DataFrame(rows)
        df_rows.to_csv(OUTPUT_FILE, index=False, encoding="utf-8-sig")
        print(f"Résultats sauvegardés : {OUTPUT_FILE} ({len(df_rows)} lignes)")
    else:
        print("Aucune ligne de résultat à sauvegarder.")

    # Sauvegarde CSV erreurs
    if errors:
        df_err = pd.DataFrame(errors)
        df_err.to_csv(ERROR_FILE, index=False, encoding="utf-8-sig")
        print(f"Erreurs sauvegardées : {ERROR_FILE} ({len(df_err)} lignes)")
    else:
        print("Aucune erreur.")

    print(f"Nombre d'appels effectués : {calls_done}")


if __name__ == "__main__":
    main()
```

```python
import requests
import csv
import time
from datetime import datetime
#api pour Prochains passages (plateforme Île-de-France Mobilités) - requête globale 
API_KEY = "L3mYGua22eGo1P5wRvN3jzyNTgj3raQA"
URL = "https://prim.iledefrance-mobilites.fr/marketplace/disruptions_bulk/disruptions/v2"

OUTPUT_FILE = "test_sncf.csv"

START_TIME = time.time()
DURATION = 60  # 1 minute

all_rows = []


def flatten_json(data, parent_key="", sep="_"):
    items = {}

    if isinstance(data, dict):
        for k, v in data.items():
            new_key = f"{parent_key}{sep}{k}" if parent_key else k
            items.update(flatten_json(v, new_key, sep))

    elif isinstance(data, list):
        for i, v in enumerate(data):
            new_key = f"{parent_key}{sep}{i}"
            items.update(flatten_json(v, new_key, sep))

    else:
        items[parent_key] = data

    return items


def fetch():
    headers = {
        "accept": "application/json",
        "apikey": API_KEY
    }
    response = requests.get(URL, headers=headers)
    response.raise_for_status()
    return response.json()


request_count = 0

while time.time() - START_TIME < DURATION:

    print("\n🚀 Nouvelle salve de 5 requêtes")

    for i in range(5):
        if time.time() - START_TIME >= DURATION:
            break

        try:
            data = fetch()
            disruptions = data.get("disruptions", [])

            print(f"Requête {request_count + 1} → {len(disruptions)} perturbations")

            for d in disruptions:
                flat = flatten_json(d)
                flat["timestamp"] = datetime.now().isoformat()
                all_rows.append(flat)

            request_count += 1

        except Exception as e:
            print("Erreur:", e)

    # pause 7 secondes
    if time.time() - START_TIME < DURATION:
        print("⏸ Pause 7 secondes...")
        time.sleep(7)


# 📝 Sauvegarde CSV
if all_rows:
    keys = set()
    for row in all_rows:
        keys.update(row.keys())

    keys = sorted(keys)

    with open(OUTPUT_FILE, "w", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=keys)
        writer.writeheader()
        writer.writerows(all_rows)

    print(f"\n✅ CSV généré: {OUTPUT_FILE}")
    print(f"📊 Total requêtes: {request_count}")
    print(f"📄 Total lignes: {len(all_rows)}")

else:
    print("❌ Aucune donnée récupérée")
```

```python
import xml.etree.ElementTree as ET
import pandas as pd
import os

# Chemin du fichier XML
file_path = r"C:\Users\User\Desktop\OFFRE_800_SNCF\calendriers.xml"

# Dossier de sortie
folder = r"C:\Users\User\Desktop\SNCF\csv"
os.makedirs(folder, exist_ok=True)

# Fichier CSV final
output_path = os.path.join(folder, "calendriers.csv")

# Charger le XML
tree = ET.parse(file_path)
root = tree.getroot()

# Namespace NeTEx
ns = {"ns": "http://www.netex.org.uk/netex"}

data = []

# Parcourir tous les DayType
for daytype in root.findall(".//ns:DayType", ns):
    day_id = daytype.get("id", "")
    created = daytype.get("created", "")
    changed = daytype.get("changed", "")
    data_source = daytype.get("dataSourceRef", "")
    version = daytype.get("version", "")

    name_elem = daytype.find("ns:Name", ns)
    name = name_elem.text.strip() if name_elem is not None and name_elem.text else ""

    # Récupérer tous les jours s'il y en a
    days_elems = daytype.findall(".//ns:DaysOfWeek", ns)
    days_list = [d.text.strip() for d in days_elems if d.text]
    days = ", ".join(days_list)

    data.append({
        "id": day_id,
        "created": created,
        "changed": changed,
        "dataSourceRef": data_source,
        "version": version,
        "name": name,
        "days": days
    })

# Créer DataFrame
df = pd.DataFrame(data)

# Export CSV
df.to_csv(output_path, index=False, encoding="utf-8-sig")

print("✅ CSV créé :", output_path)
print(df.head())
print("Nombre de lignes :", len(df))
```

```python
import requests
import csv
import time
from datetime import datetime


def prochains_passages(line_ref: str, api_key: str):
    url = "https://prim.iledefrance-mobilites.fr/marketplace/requete-ligne"

    headers = {
        "apikey": api_key
    }

    params = {
        "LineRef": line_ref
    }

    response = requests.get(url, headers=headers, params=params, timeout=30)
    response.raise_for_status()
    return response.json()


def flatten_json(obj, prefix=""):
    resultat = {}

    if isinstance(obj, dict):
        for cle, valeur in obj.items():
            nouvelle_cle = f"{prefix}{cle}" if prefix == "" else f"{prefix}_{cle}"
            resultat.update(flatten_json(valeur, nouvelle_cle))

    elif isinstance(obj, list):
        for i, valeur in enumerate(obj):
            nouvelle_cle = f"{prefix}_{i}" if prefix else str(i)
            resultat.update(flatten_json(valeur, nouvelle_cle))

    else:
        resultat[prefix] = obj

    return resultat


def extraire_passages_complets(data):
    lignes = []

    try:
        deliveries = data.get("Siri", {}).get("ServiceDelivery", {}).get("EstimatedTimetableDelivery", [])

        for delivery in deliveries:
            frames = delivery.get("EstimatedJourneyVersionFrame", [])

            for frame in frames:
                vehicles = frame.get("EstimatedVehicleJourney", [])

                for vehicle in vehicles:
                    calls = vehicle.get("EstimatedCalls", {}).get("EstimatedCall", [])

                    if calls:
                        for call in calls:
                            enregistrement = {
                                "date_requete": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
                                "vehicle": vehicle,
                                "call": call
                            }
                            lignes.append(flatten_json(enregistrement))
                    else:
                        enregistrement = {
                            "date_requete": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
                            "vehicle": vehicle
                        }
                        lignes.append(flatten_json(enregistrement))

    except Exception as e:
        print("Erreur lors de l'extraction :", e)

    return lignes


def lire_entete_csv(fichier):
    try:
        with open(fichier, mode="r", newline="", encoding="utf-8") as f:
            reader = csv.reader(f)
            return next(reader, [])
    except FileNotFoundError:
        return []


def lire_lignes_csv(fichier):
    try:
        with open(fichier, mode="r", newline="", encoding="utf-8") as f:
            reader = csv.DictReader(f)
            return list(reader)
    except FileNotFoundError:
        return []


def sauvegarder_csv_auto(data, fichier="passages_3_heures_complet.csv"):
    if not data:
        print("Aucune donnée à sauvegarder.")
        return

    anciennes_colonnes = lire_entete_csv(fichier)
    anciennes_lignes = lire_lignes_csv(fichier)

    nouvelles_colonnes = set()
    for ligne in data:
        nouvelles_colonnes.update(ligne.keys())

    toutes_les_colonnes = list(dict.fromkeys(anciennes_colonnes + sorted(nouvelles_colonnes)))

    if anciennes_colonnes and set(toutes_les_colonnes) != set(anciennes_colonnes):
        for ancienne_ligne in anciennes_lignes:
            for col in toutes_les_colonnes:
                if col not in ancienne_ligne:
                    ancienne_ligne[col] = ""

        with open(fichier, mode="w", newline="", encoding="utf-8") as f:
            writer = csv.DictWriter(f, fieldnames=toutes_les_colonnes)
            writer.writeheader()
            writer.writerows(anciennes_lignes)

    mode_fichier = "a" if anciennes_lignes or anciennes_colonnes else "w"

    with open(fichier, mode=mode_fichier, newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=toutes_les_colonnes)

        if mode_fichier == "w":
            writer.writeheader()

        for ligne in data:
            ligne_complete = {col: ligne.get(col, "") for col in toutes_les_colonnes}
            writer.writerow(ligne_complete)

    print(f"Données sauvegardées dans : {fichier}")


def collecter_pendant_2_minutes(line_ref: str, api_key: str, intervalle=10, duree=120):
    debut = time.time()
    fichier = "passages_3_heures_complet.csv"

    while time.time() - debut < duree:
        try:
            heure_actuelle = datetime.now().strftime("%H:%M:%S")
            print(f"Requête envoyée à {heure_actuelle}...")

            data = prochains_passages(line_ref, api_key)
            lignes = extraire_passages_complets(data)

            print(f"{len(lignes)} lignes extraites.")
            sauvegarder_csv_auto(lignes, fichier)

        except requests.exceptions.HTTPError as e:
            print(f"Erreur HTTP : {e}")
            if e.response is not None:
                print("Détail :", e.response.text)

        except requests.exceptions.RequestException as e:
            print(f"Erreur réseau : {e}")

        except Exception as e:
            print(f"Erreur inattendue : {e}")

        time.sleep(intervalle)

    print(f"Collecte terminée. Fichier final : {fichier}")


if __name__ == "__main__":
    API_KEY = "L3mYGua22eGo1P5wRvN3jzyNTgj3raQA"
    LINE_REF = "STIF:Line::C01742:"  # Exemple RER A

    collecter_pendant_2_minutes(
        line_ref=LINE_REF,
        api_key=API_KEY,
        intervalle=10,   # une requête toutes les 10 secondes
        duree=3600        # pendant 2 minutes
    )
```
